### Categorical Variables

```python
ax = sns.countplot(data=adult, y='workclass')

initialy=0
for p in ax.patches: # each bar and its background area is called a patch
    ax.text(p.get_width(), initialy+p.get_height()/8, s="{:.2f}%".format(p.get_width()/len(adult)*100))
    initialy+=1

plt.tight_layout()
```

```python
def barchart(df, attribute):
    ax = sns.countplot(data=df, y=attribute, order= adult[attribute].value_counts().index)

    initialy=0
    for p in ax.patches: # each bar and its background area is called a patch
        ax.text(p.get_width(), initialy+p.get_height()/8, s="{:.2f}%".format(p.get_width()/len(df)*100))
        initialy+=1

    plt.tight_layout()
```

### Race and Income

```python
sns.displot(adult, x='race', hue='income', multiple='stack', height=10, aspect=1.5)
plt.tight_layout()
plt.show()
```

```python
black = adult[adult['race']=='Black']
black_percent = len(black[black['income']==1]) / len(black) * 100
print("{:.2f}% of black americans make over 50k".format(black_percent))

white = adult[adult['race']=='White']
white_percent = len(white[white['income']==1]) / len(white) * 100
print("{:.2f}% of white americans make over 50k".format(white_percent))

asian = adult[adult['race']=='Asian-Pac-Islander']
asian_percent = len(asian[asian['income']==1]) / len(asian) * 100
print("{:.2f}% of asian americans make over 50k".format(asian_percent))
```

From the stacked barplot, we can see that that race plays a factor in determining income. The stacked barplot allows us to interpret race and income proportionally. We can see that out of those who identify as white, roughly a quarter of them make over 50k. The same is true for Asians, around 27% of Asians. On the other hand, only 12% of Black Americans make over 50k. Thus, an individual who is Asian or White is more than twice as likely within their respective racial groups to make over 50k than an individual who is Black.

### Immigrants and Income
Pitching immigrants vs US-born Americans is a narrative spewed by the media and politicians, often for political purposes. Regardless of the reasons for this narrative, is there any truth to it? Are immigrants really "stealing" higher paying jobs from  US-born Americans?

Let's explore the data.

```python
# Subset native-country column into groups: United-States and Immigrants
# Create new immigrant column
adult_immigrant = adult.copy()
adult_immigrant.loc[(adult_immigrant['native-country']!='United-States'), 'native-country'] = 'Immigrant'
```

```python
adult_immigrant['native-country'].value_counts()
```

```python
plt.figure(figsize=[12,10])
sns.countplot(data=adult_immigrant, x='native-country', hue='income')
plt.tight_layout()
```

```python
immigrant = adult[adult_immigrant['native-country']=='Immigrant']
usa = adult[adult_immigrant['native-country']=='United-States']
```

```python
len(immigrant) / len(adult) * 100
```

Immmigrants make up roughly 10 percent of the total sample population

```python
population_immigrant_1 = len(immigrant[immigrant['income']==1]) / len(adult_immigrant) * 100
immigrant_1 = len(immigrant[immigrant['income']==1]) / len(immigrant) * 100
usa_1 = len(usa[usa['income']==1]) / len(usa) * 100 

print("The number of immigrants with incomes over 50k is only {:.2f}% of the total sample population"
      .format(population_immigrant_1))
print("Of the immigrant population, which makes up roughly 11% of the total sample population, {:.2f}% make over 50k"
     .format(immigrant_1))
print("Of the US-native population, which makes up roughly 89% of the total sample population, {:.2f}% make over 50k"
     .format(usa_1))
```

**Interpretation:**

The data shows that immigrants are a small minority of the total American population. Most immigrants are not threatening high-income jobs in the USA. In fact, on top of the overall high-income, immigrant population being smaller than the US-born population, the proportion of that that small immigrant population with high-income jobs is less than the propotion of high-income jobs for Americans.
